```python
pdf_file = "/content/resumes/25797445.pdf" # resume file path
```

```python
import re
import PyPDF2

pdf_reader = PyPDF2.PdfReader(pdf_file)
text = ""
for page in pdf_reader.pages:
    text += page.extract_text()

    category = None
    skills = []
    education = []

    # Extract the category
    category = text.split("\n")[0]

    # Extract the skills
    # ----------------using resume parser-----------------------

    from resume_parser import resumeparse
    data = resumeparse.read_file(pdf_file)

    skills = data['skills']

    # ----------------USING REGEX-----------------------------
    # match = re.findall(r"Skills\s(.*)", text)
    # if match:
    #     skills_list = match[0].split(", ")
    #     for skill in skills_list:
    #         skills.append(skill.strip())

    # Extract the education
    match = re.findall(r"(?:Education(?: and Training)?)([\s\S]+?)(?:Skills|Summary|Accomplishments|$)", text)
    for education_details in match:
        education.append(education_details.replace("\n",""))                         # list
        # education = ""
        # education = education_details                             # str

# print(text)
# print(category)
# print(skills)
# print(education)

cv_details = []
cv_details.append(category)
cv_details.append(skills)
cv_details.append(education)
print(cv_details)
```
